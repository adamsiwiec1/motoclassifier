```python
import matplotlib.pyplot as plt
import json
import numpy as np
import os
```

```python
def compare_accuracy(data, ax, file_path):
    ax.set_ylim(0, 1)
    ax.set_yticks(np.arange(0, 1, 0.1))
    ax.plot(data['accuracy'], label="accuracy")
    ax.set_title(file_path.split('/')[-1].split('.')[0])
    ax.plot(data['val_accuracy'], label="val_accuracy")

def compare_auc(data, ax, file_path):
    ax.plot(data['auc'], label="auc")
    ax.set_title(file_path.split('/')[-1].split('.')[0])
    ax.plot(data['val_auc'], label="val_auc")

def compare_mse(data, ax, file_path):
    ax.plot(data['mean_squared_error'], label="mse")
    ax.set_title(file_path.split('/')[-1].split('.')[0])
    ax.plot(data['val_mean_squared_error'], label="val_mse")

def compare_loss(data, ax, file_path):
    ax.plot(data['loss'], label="loss")
    ax.set_title(file_path.split('/')[-1].split('.')[0])
    ax.plot(data['val_loss'], label="val_loss")

def compare_history(history_files):
    plots = [compare_accuracy, compare_auc, compare_mse, compare_mse]
    for plot in plots:
        fig, axs = plt.subplots(1, len(history_files), figsize=(15, 3))
        for i, ax in enumerate(axs):
            file_path = history_files[i]
            with open(file_path, "r") as file:
                data = json.load(file)
            plot(data, ax, file_path)
        plt.legend()
        plt.show()
        

compare_history(["./archive/colin_adadelta-history.json", './archive/colin_nadam-history.json'])
```
